```python
#import needed libraries
from sqlalchemy import create_engine
import pandas as pd
import os
import time
#get password from environmnet var
pwd = 'password'
uid = 'user'
server = "localhost"
db = "deteccao_fraude"
port = "5432"
#
engine = create_engine(f'postgresql://{uid}:{pwd}@{server}:{port}/{db}')
```

```python
df = pd.read_csv(r"C:\Users\Matheus\Desktop\Estudo\Kafka\Transação_Credito\Previsao\Dados\card_transdata.csv")
```

```python
#Deixando o mesmo numero de categorias da variável fraudes para uma melhor visualização
fraudes = df[df['fraud'] == 1].head(50)
n_fraudes = df[df['fraud'] == 0].head(50)

dados = pd.concat([fraudes, n_fraudes], ignore_index=True)

# Embaralhar as linhas
dados = dados.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
dados = dados.drop('fraud', axis = 1)
```

```python
#Inserindo dados no Postgres
for index, row in dados.iterrows():
    mod = pd.DataFrame(row.to_frame().T)
    mod.to_sql(f"transacoes", engine, if_exists='append', index=False)
    print(f"Row Inserted {index}")
    time.sleep(3)
```
